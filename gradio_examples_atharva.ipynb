```python
import gradio as gr

def predict_wine_class(alcohol, malic_acid, ash, alcalinity_of_ash, magnesium, total_phenols, flavanoids, nonflavanoid_phenols, proanthocyanins, color_intensity, hue, od280_od315, proline):
    """
    Predict wine class using the trained model.
    """
    # Convert the inputs to a DataFrame
    input_data = pd.DataFrame([[alcohol, malic_acid, ash, alcalinity_of_ash, magnesium, total_phenols, flavanoids, nonflavanoid_phenols, proanthocyanins, color_intensity, hue, od280_od315, proline]],
                              columns=wine.feature_names)

    # Predict with the model
    prediction = predict_model(final_model, data=input_data)
    predicted_class = wine.target_names[prediction['Label'].values[0]]

    return predicted_class

# Define Gradio interface components
inputs = [
    gr.inputs.Slider(minimum=10, maximum=15, step=0.1, default=13, label="Alcohol"),
    gr.inputs.Slider(minimum=0, maximum=5, step=0.1, default=2, label="Malic acid"),
    gr.inputs.Slider(minimum=1, maximum=3, step=0.1, default=2, label="Ash"),
    gr.inputs.Slider(minimum=10, maximum=30, step=0.1, default=20, label="Alcalinity of ash"),
    gr.inputs.Slider(minimum=70, maximum=140, step=1, default=100, label="Magnesium"),
    gr.inputs.Slider(minimum=0, maximum=4, step=0.1, default=2, label="Total phenols"),
    gr.inputs.Slider(minimum=0, maximum=4, step=0.1, default=2, label="Flavanoids"),
    gr.inputs.Slider(minimum=0, maximum=1, step=0.1, default=0.5, label="Nonflavanoid phenols"),
    gr.inputs.Slider(minimum=0, maximum=3, step=0.1, default=1, label="Proanthocyanins"),
    gr.inputs.Slider(minimum=0, maximum=13, step=0.1, default=6, label="Color intensity"),
    gr.inputs.Slider(minimum=0, maximum=2, step=0.1, default=1, label="Hue"),
    gr.inputs.Slider(minimum=1, maximum=4, step=0.1, default=2.5, label="OD280/OD315 of diluted wines"),
    gr.inputs.Slider(minimum=300, maximum=1800, step=10, default=1000, label="Proline")
]

outputs = gr.outputs.Label()

# Launch the Gradio interface
gr.Interface(fn=predict_wine_class, inputs=inputs, outputs=outputs, live=True).launch()
```

```python
import gradio as gr
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target

# Train a simple RandomForest model
clf = RandomForestClassifier()
clf.fit(X, y)

def predict_species(sepal_length, sepal_width, petal_length, petal_width):
    pred = clf.predict([[sepal_length, sepal_width, petal_length, petal_width]])
    return iris.target_names[pred[0]]

# Gradio UI
iface = gr.Interface(fn=predict_species,
                     inputs=["number", "number", "number", "number"],
                     outputs="text")
iface.launch()
```

```python
import gradio as gr
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Load dataset
digits = datasets.load_digits()

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.5, shuffle=False)

# Train a simple RandomForest model
clf = RandomForestClassifier(n_estimators=1000)
clf.fit(X_train, y_train)

def recognize_digit(arr):
    pred = clf.predict([arr])
    return str(pred[0])

# Gradio UI
iface = gr.Interface(fn=recognize_digit,
                     inputs=gr.inputs.Image(shape=(8, 8), image_mode='L', invert_colors=True),
                     outputs="text")
iface.launch()
```
